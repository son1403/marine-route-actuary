```python
import pandas as pd
from shapely.geometry import Polygon
from marine_route_actuary import flag_high_risk_routes
```

## Step 2: Create example data

We create a small DataFrame with two routes. Place names can be English or Vietnamese.


```python
df = pd.DataFrame(
    [
        {"FROM": "Singapore", "TO": "Dubai"},
        {"FROM": "Hồ Chí Minh City", "TO": "Hong Kong"},
    ]
)
df
```

## Step 3: Define coordinates and a high-risk area

We provide a lookup table (place -> lon/lat) and a simple polygon.


```python
lookup = {
    "Singapore": (103.8198, 1.3521),
    "Dubai": (55.2708, 25.2048),
    "Ho Chi Minh City": (106.6297, 10.8231),
    "Hong Kong": (114.1694, 22.3193),
}

# Example high-risk area polygon (toy example)
risk_area = Polygon([
    (45.0, 10.0),
    (60.0, 10.0),
    (60.0, 30.0),
    (45.0, 30.0),
])
```

## Step 4: Flag high-risk routes

This adds a boolean `high_risk` column to the DataFrame.


```python
out = flag_high_risk_routes(
    df,
    location_lookup=lookup,
    route_engine="scgraph",  # default, maritime routing
    high_risk_areas=[risk_area],
)
out
```

## Pitfalls and Extensions

- Pitfall: unresolved place names silently produce `high_risk=False`.
  Always QA unresolved names or provide a complete lookup table.
- Extension: replace the toy polygon with your real high-risk area geometry
